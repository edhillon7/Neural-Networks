```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
inputs = iris.data
inputs
```

```python
outputs = iris.target 
outputs
```

```python
iris.feature_names
```

```python
#define train and test dataset 

X_train, X_test, Y_train, Y_test = train_test_split(inputs, outputs, test_size = 0.2)
print (X_train)
```

```python
#making the network(default number of neurons in hidden layer: 100)
#training of the neural network 
#we can do various things 
network = MLPClassifier(max_iter= 2000, verbose = True, tol=0.0000100, solver = 'adam', activation = 'logistic', learning_rate = 'constant', learning_rate_init = 0.001, batch_size = 32, early_stopping = True, n_iter_no_change = 50)
network.fit(X_train, Y_train)


network.fit(X_train, Y_train)
```
